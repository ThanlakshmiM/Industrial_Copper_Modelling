```python
import pandas as pd
import numpy as np
def clean_dataset(columns):
    dff=df[columns]
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    dff[columns].dropna(inplace=True)
    indices_to_keep = ~dff[columns].isin([np.nan, np.inf, -np.inf]).any(axis=1)
    return dff[indices_to_keep].astype(np.float64)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
def plot(columns):
  for i in columns:
     plt.figure(figsize=(16,5))
     plt.subplot(1,2,1)
     sns.distplot(df1[i])
     plt.subplot(1,2,2)
     sns.boxplot(data=df1,x=i)
     plt.show()
```

after Treating Outliers

```python
import numpy as np
def treat_outliers(columns):
    for i in columns:
        percentile25 = df1[i].quantile(0.25)
        percentile75 = df1[i].quantile(0.75)
        iqr = percentile75 - percentile25      # IQR = Q3 - Q1
        upper_limit = percentile75 + 1.5 * iqr
        lower_limit = percentile25 - 1.5 * iqr
        df1[i]= np.where(df1[i] > upper_limit, upper_limit, np.where(df1[i] < lower_limit, lower_limit, df1[i]))
    print('successful')
treat_outliers(columns)
```

Comparing Models

Let's compare the train,test and accuracy score of all the classifiers models used above.

```python
# demonstration of calculating metrics for a neural network model using sklearn
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import cohen_kappa_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
```

```python
def all_algorithms_classifiers_score(x_train,y_train,x_test,y_test):
  score={'Model':[],'Train Score':[],'Test Score':[],'Accuracy':[],'Precision':[],'Recall':[],'F1 Score':[],'Kappa':[],'ROC AUC':[]}
  classifier=['DecisionTreeClassifier','ExtraTreesClassifier','RandomForestClassifier','XGBClassifier','KNeighborsClassifier']
  for i in range(len(classifier)):
    cl=algorithm(classifier[i])
   # new_X = np.nan_to_num(x_train.astype(np.float32))
    cl.fit(x_train,y_train)
    y_pred=cl.predict(x_test)
    train_score = str(round((cl.score(x_train, y_train))* 100, 2)) + '%'
    test_score = str(round((cl.score(x_test, y_test))* 100, 2)) + '%'
    accuracy= str(round((accuracy_score(y_pred,y_test))* 100, 2)) + '%'
    accuracy = str(round((accuracy_score(y_test, y_pred))* 100, 2)) + '%'
    precision = str(round((precision_score(y_test, y_pred))* 100, 2)) + '%'   #Precision = true_positives / (true_positives + false_positives)
    recall = str(round((recall_score(y_test, y_pred))* 100, 2)) + '%'          #Recall = true_positives / (true_positives + false_negatives)
    f1 = str(round((f1_score(y_test, y_pred))* 100, 2)) + '%'                    # F1 Score = 2 * (precision * recall) / (precision + recall)
    kappa = str(round((cohen_kappa_score(y_test, y_pred))* 100, 2)) + '%'
    auc = str(round((roc_auc_score(y_test, y_pred))* 100, 2)) + '%'             # ROC AUC = Area Under ROC Curve
    score['Model'].append(classifier[i])
    score['Train Score'].append(train_score)
    score['Test Score'].append(test_score)
    score['Accuracy'].append(accuracy)
    score['Precision'].append(precision)
    score['Recall'].append(recall)
    score['F1 Score'].append(f1)
    score['Kappa'].append(kappa)
    score['ROC AUC'].append(auc)

  return score
```

```python
def regression_algorithm(algorithm):

    if algorithm == 'ExtraTreesRegressor':
        return ExtraTreesRegressor()
    elif algorithm == 'DecisionTreeRegressor':
        return DecisionTreeRegressor()
    elif algorithm == 'LinearRegression':
        return LinearRegression()
    elif algorithm == 'RandomForestRegressor':
        return RandomForestRegressor()
    elif algorithm == 'GradientBoostingRegressor':
        return GradientBoostingRegressor()
    else:
        print('Error')
```

```python
import numpy as np
import pickle
alg=['ExtraTreesRegressor','DecisionTreeRegressor','LinearRegression','RandomForestRegressor','GradientBoostingRegressor']
input_data=np.array([[54.151139,30156308.0,28.0,7,5,10.0,2.00,1500.0,5384,1670798778,2021,4,1,2021,7,1]])
File=[]
for i in alg:
       model = regression_algorithm(i)
       model.fit(scaler_trans, y_train)
       with open(f'{i}.pkl', 'wb') as file:
          pickle.dump(model, file)
          File.append(f'{i}.pkl')
       with open(f'{i}.pkl', 'rb') as file:
          result = pickle.load(file)
       scaled_new_data_point = scaler.transform(input_data)
       predicted_output = result.predict(scaled_new_data_point)
       print(f'{i},scaler : {predicted_output[0]}')

print('Created Pickle File',*File)
```
